```python
import pandas as pd

df = pd.read_csv("SpotifyData.csv")
df
```

```python
# Dropping columns for variables not considered.

col = ['acousticness', 'speechiness', 'valence', 'tempo', 'duration_ms', 'release_date', 'year']
df1 = df[col]
df1
```

```python
# Filtering and removing rows to limit to releases in the last 40 years (1980-2020).

df2 = df1[(df1["year"] > 1979)]
df2
```

```python
df2.to_csv("clean_data.csv", index=False)
df2
```

```python
df3 = pd.read_csv("clean_data.csv")
df3
```

```python
print(df3['release_date'].dtype)
```

```python
# Removing rows without a month indicated in the release_date column.

df3['release_date'] = df3['release_date'].astype(str)
df3 = df3.loc[df3['release_date'].str.len () == 10]
df3
```

```python
df3.to_csv("clean_data2.csv", index=False)
df3
```

```python
df4 = pd.read_csv("clean_data2.csv")
df4
```

```python
# Create new column with just the month (taken from release_date) and cosmetic fix to duration column title 
# and changing back data type for release_date.

df4['month'] = df4['release_date'].str[5:7]
df4['month'] = df4['month'].astype(int)
df4['release_date'] = df4['release_date'].astype(object)
df4
```

```python
df4.to_csv("cleaned_data.csv", index=False)
df4
```

```python
df5 = pd.read_csv("cleaned_data.csv")
df5
```

```python
# Getting averages of features by month and year.

avg_extrap = df5.groupby(['year', 'month']).agg(
    avg_acousticness = ('acousticness', 'mean'),
    avg_duration = ('duration_ms', 'mean'),
    avg_speechiness = ('speechiness', 'mean'),
    avg_valence = ('valence', 'mean'),
    avg_tempo = ('tempo', 'mean')
)
print(avg_extrap)
```

```python
avg_extrap.to_csv("averages_year_month.csv")
avg_extrap
```
